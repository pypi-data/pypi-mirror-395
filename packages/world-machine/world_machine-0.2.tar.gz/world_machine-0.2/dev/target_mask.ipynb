```python
import torch
```

```python
target = torch.zeros([1,100, 3])
target[0, 50:] = 1

logits = torch.zeros([1,100, 3])
logits[0, 25:50] = 1
logits[0, 75:100] = 1
```

```python
criterion = torch.nn.MSELoss()
```

```python
criterion(target,logits)
```

```python
mask = torch.ones([1,100, 3], dtype=bool)
mask[0, :25, 0] = False
mask[0, 75:, 0] = False
```

```python
criterion(target*mask,logits*mask)
```

```python
criterion(target[mask],logits[mask])
```

```python
criterion(target*mask,logits*mask)*(mask.numel()/mask.sum())
```
