```python
import torch
```

```python
t = torch.ones([32,100,2])
```

```python
end = 200
```

```python
if t.shape[1] != 200:
    expansion = torch.empty([t.shape[0], end-t.shape[1], t.shape[2]], dtype=t.dtype, device=t.device)
    t = torch.hstack([t, expansion])
```

```python
t[0,:,0]
```

```python
device=t.device
```

```python
from tensordict import TensorDict
import tensordict
```

```python
sensorial_data = TensorDict(batch_size=[32,100])
sensorial_data["a"] = torch.ones([32, 100, 3])
```

```python
sensorial_data.keys()
```

```python
sensorial_data["a"].shape[2:]
```

```python
sensorial_data.batch_size
```

```python
if sensorial_data.batch_size[1] != 200:
    expansion_seq_len = 100
    expansion = TensorDict(device=device, batch_size=[32,expansion_seq_len])

    for key in sensorial_data.keys():
        expansion[key] = torch.empty([32, expansion_seq_len]+list(sensorial_data["a"].shape[2:]), device=device)

    sensorial_data = torch.cat([sensorial_data, expansion], dim=1)
```
