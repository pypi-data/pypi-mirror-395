```python
import torch
import numpy as np
```

```python
n_past = 5
```

```python
data = torch.arange(0,100,1).repeat(32).reshape(32, 100)
data
```

```python
mask = torch.ones(100, dtype=bool)
```

```python
for i in range(1):
    past_index = i+2

    past_data = torch.roll(data, -past_index, 1)
    past_mask = torch.roll(mask, -past_index)

    past_mask[past_mask.shape[0]-i:] = False
```

```python
print(past_data[0,:10])
print(data[0,:10])
print(past_mask[:10].int())
```

```python
print(past_data[0,90:])
print(data[0,90:])
print(past_mask[90:].int())
```
