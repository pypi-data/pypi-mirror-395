```python
import numpy as np
def generator_numpy(seed:int|list[int], function) -> np.random.Generator:
    if isinstance(seed, int):
        seed = [seed]

    seed = [int.from_bytes(bytes(function.__qualname__, "utf-8"))]+seed
    
    return np.random.default_rng(seed=seed)
```
