```python
import zipfile
import pathlib
import os

import tqdm
```

```python
base_dir = pathlib.Path("test")
```

```python
base_dir = pathlib.Path(base_dir)
```

```python
list(base_dir.rglob("*"))
```

```python
with zipfile.ZipFile("directory_tree.zip", 
                     mode="w", 
                     compression=zipfile.ZIP_LZMA) as archive:
    for file_path in tqdm.tqdm(base_dir.rglob("*")):
        if "CS" in str(file_path):
            continue

        if os.path.isfile(file_path):
            if str(file_path).endswith(".npz"):
                archive.write(
                file_path,
                arcname=file_path.relative_to(base_dir)
             )
        else:
            archive.write(
                file_path,
                arcname=file_path.relative_to(base_dir)
            )
```
