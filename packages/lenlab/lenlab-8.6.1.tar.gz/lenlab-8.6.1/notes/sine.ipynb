```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def interval_25ns(f):
    if f < 250.0:
        return 200
    if f < 500.0:
        return 80
    return 40
```

```python
frequencies = np.logspace(2, 4, 201)
intervals_25ns = np.vectorize(interval_25ns)(frequencies)
```

```python
points = 2 * np.round(1 / (25e-9 * intervals_25ns * frequencies * 2))  # even numbers
```

```python
fig, ax = plt.subplots()
ax.semilogx(frequencies, points)
ax.grid()
```

```python
print("sine_table = [  # frequency_hertz, interval_25ns, points")
for p, i in zip(points, intervals_25ns, strict=True):
    print(f"    ({int(round(40e6 / i / p))}, {int(round(i))}, {int(round(p))}),")
print("]")
```
