```python
import glob
import os
import json
import shutil

import hashlib
from world_machine_experiments.shared.save_parameters import make_model
```

```python
keys = ['n_segment',
 'fast_forward',
 'stable_state_epochs',
 'check_input_masks',
 'state_save_method',
 'mask_sensorial_data',
 'short_time_recall',
 'recall_stride_past',
 'recall_stride_future',
 'recall_n_past',
 'recall_n_future',
 'positional_encoder_type',
 'block_configuration',
 'state_activation',
 'state_regularizer',
 'train_mse',
 'train_sdtw',
 'noise_config']
```

```python
base_dir = "toy1d_experiment1_configuration_test"
variation_paths = glob.glob(os.path.join(base_dir,"*"))
```

```python
for i in range(len(variation_paths)):
    try:
        path = variation_paths[i]

        parameters_path = os.path.join(path, "parameters.json")

        with open(parameters_path) as file:
            parameters = json.load(file)

        config = {}

        for key in keys:
            config[key] = parameters["parameters"][key]

        model = make_model(config, "ParametersModel").model_validate(config)
        model_json = model.model_dump_json()
        variation_hash = int(
            hashlib.md5(model_json.encode('utf-8')).hexdigest(), 16)

        new_name = "variation"+str(variation_hash)
        new_path = os.path.join(base_dir, new_name)

        os.rename(path, new_path)
    
    except Exception as e:
        print(i, "with error:")
        print(e)
        print("-------")
```
