# `sierpinski_foam`

The sierpinski gasket is a classic example of a fractal image with self-similar properties across all length scales.  It has also been extensively analyzed so these properties are well know.  The ``sierpinski_foam`` is a 3D version of the gasket, but also produces 2D images.

```python
import matplotlib.pyplot as plt

import porespy as ps
```

## `shape`

The original version of this generator did not accept shape, and instead its size was dictated by the number of divisions requested.  The new version does except shape and instead it truncates the pattern if the requested number of divisions exceeds the shape. 

In the images below, the image of size 100 by 100 is not symmetrical.  If the full pattern is desired, then the shape must be set to `3**n` in all directions, as shown on the right (`n=5` is the default value).

```python
fig, ax = plt.subplots(1, 2, figsize=[12, 6])

im = ps.generators.sierpinski_foam(shape=[3**4, 3**5], n=3)
ax[0].imshow(im, interpolation="none")
ax[0].axis(False)

im = ps.generators.sierpinski_foam(shape=[3**5, 3**5], n=5)
ax[1].imshow(im, interpolation="none")
ax[1].axis(False);
```

## `n`

The number of times the pattern is divided is controlled by the `n` parameter:

```python
n = 3
im1 = ps.generators.sierpinski_foam(shape=[3**n, 3**n], n=n)

n = 4
im2 = ps.generators.sierpinski_foam(shape=[3**n, 3**n], n=n)

fig, ax = plt.subplots(1, 2, figsize=[12, 6])
ax[0].imshow(im1, interpolation="none")
ax[0].axis(False)

ax[1].imshow(im2, interpolation="none")
ax[1].axis(False);
```

## `mode`

If the pattern and the shape of the image are not in agreement the pattern will be truncated. How this is applied is controlled by the `mode` argument with options being `'upper'`, `'centered'`, and `None`:

```python
im1 = ps.generators.sierpinski_foam(shape=[100, 100], n=n, mode=None)
im2 = ps.generators.sierpinski_foam(shape=[100, 100], n=n, mode="upper")
im3 = ps.generators.sierpinski_foam(shape=[100, 100], n=n, mode="centered")

fig, ax = plt.subplots(1, 3, figsize=[12, 4])
ax[0].imshow(im1, interpolation="none")
ax[0].axis(False)

ax[1].imshow(im2, interpolation="none")
ax[1].axis(False)

ax[2].imshow(im3, interpolation="none")
ax[2].axis(False);
```
